```python
import pandas as pd
import numpy as np
```

## Extract inventario diario

```python
df_inventario_diario = pd.read_csv('../data/fotografia_inventario_diario_sanitizado.csv')
display(df_inventario_diario.head(20))
display(df_inventario_diario.info())
```

#### ELiminar duplicados exactos

```python
df_inventario_diario = df_inventario_diario.drop_duplicates(subset=['fecha', 'id_producto', 'stock_cierre' ,'almacen'], keep='first')
display(df_inventario_diario.head())
```

#### Estandarizacion de fecha


```python
df_inventario_diario["fecha"] = pd.to_datetime(df_inventario_diario["fecha"], format="mixed")
display(df_inventario_diario.info()) 
```

#### Estandarizacion texto

```python
df_inventario_diario['id_producto'] = df_inventario_diario['id_producto'].str.replace(' ', '').str.strip().str.upper()
df_inventario_diario['almacen'] = df_inventario_diario['almacen'].str.replace(' ', '').str.strip().str.upper()
display(df_inventario_diario.head())
```

#### Coercion numerica

```python
df_inventario_diario['stock_cierre'] = pd.to_numeric(df_inventario_diario['stock_cierre'], errors='coerce')
df_inventario_diario['stock_cierre'] = df_inventario_diario['stock_cierre'].fillna(0).astype(int)
df_inventario_diario.loc[df_inventario_diario['stock_cierre'] < 0, 'stock_cierre'] = 0
display(df_inventario_diario.head())
```

## Extract ventas diarias

```python
df_ventas_diarias = pd.read_csv('../data/ventas_diarias_sanitizado.csv')
display(df_ventas_diarias.head())
display(df_ventas_diarias.info())
```

#### Eliminar duplicados

```python
df_ventas_diarias.drop_duplicates(subset=['fecha', 'id_producto', 'cantidad_vendida', 'cliente_b2b', 'estado_orden'], keep='first')
```

#### Estandarizacion fecha

```python
df_ventas_diarias['fecha'] = pd.to_datetime(df_ventas_diarias['fecha'], format='mixed')
df_ventas_diarias.info()
```

#### Estandarizacion texto

```python
df_ventas_diarias['id_producto'] = df_ventas_diarias['id_producto'].str.upper().str.replace(r'\s+', '', regex=True)
df_ventas_diarias.head()
```

```python
df_ventas_diarias['cliente_b2b'] = df_ventas_diarias['cliente_b2b'].str.upper().str.replace(r'\s\s+', '', regex=True).str.strip()
df_ventas_diarias.head(20)
```

#### Conercion numerica

```python
df_ventas_diarias['cantidad_vendida'] = pd.to_numeric(df_ventas_diarias['cantidad_vendida'], errors='coerce')
df_ventas_diarias['cantidad_vendida'] = df_ventas_diarias['cantidad_vendida'].fillna(0).astype(int)
df_ventas_diarias.loc[df_ventas_diarias['cantidad_vendida'] < 0, 'cantidad_vendida'] = 0
df_ventas_diarias.head(20)
```

#### Merge ventas e inventario

```python
display(df_ventas_diarias.info())
display(df_inventario_diario.info())
```

```python
df_merge = pd.merge(df_ventas_diarias, df_inventario_diario, on=["fecha", "id_producto"], how="left")
display(df_merge.head(20))
```

```python
condicion_ventas = df_merge["cantidad_vendida"] == 0
condicion_stock = (df_merge["stock_cierre"] <= 0) | (df_merge["stock_cierre"].isna())
df_merge["quiebre_stock"] = np.where(condicion_ventas & condicion_stock, 1, 0)
display(df_merge.head(20))
```

```python
df_merge["ventas_sanas"] = np.where(df_merge["quiebre_stock"] == 1, np.nan, df_merge["cantidad_vendida"])
display(df_merge.head(20))
```

```python
promedio_historico = df_merge.groupby("id_producto")["ventas_sanas"].transform("mean").fillna(0).round().astype(int)
display(promedio_historico.head(20))
```

```python
df_merge["demanda_no_restringida"] = np.where(df_merge["quiebre_stock"] == 1, promedio_historico, df_merge["ventas_sanas"])
df_merge = df_merge.drop(columns=["ventas_sanas"])
display(df_merge[["fecha", "id_producto", "cantidad_vendida", "quiebre_stock", "demanda_no_restringida"]].head(20))
```
